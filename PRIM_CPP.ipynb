# PRIM - Cay khung nho nhat 

````c++
#include <iostream>
using namespace std;

const int MAX = 1001;
const int INF = 1000000000; // gia tri vo cung lon
int a[MAX][MAX];            // ma tran trong so
bool used[MAX];             // danh dau dinh da chon vao cay
int d[MAX];                 // khoang cach nho nhat tu cay den dinh v
int trace[MAX];             // luu dinh truoc do trong cay
int n, m;                   // so dinh va so canh

int main() {
    cout << "Nhap so dinh va so canh: ";
    cin >> n >> m;

    // khoi tao ma tran trong so
    for (int i = 1; i <= n; i++)
        for (int j = 1; j <= n; j++)
            a[i][j] = 0;

    // nhap danh sach canh
    for (int i = 0; i < m; i++) {
        int u, v, w;
        cin >> u >> v >> w;
        a[u][v] = a[v][u] = w; // do thi vo huong
    }

    // khoi tao ban dau
    for (int i = 1; i <= n; i++) {
        used[i] = false;
        d[i] = INF;
        trace[i] = -1;
    }

    d[1] = 0;
    int total = 0;

    for (int i = 1; i <= n; i++) {
        int u = -1, minD = INF;
        for (int j = 1; j <= n; j++) {
            if (!used[j] && d[j] < minD) {
                minD = d[j];
                u = j;
            }
        }
        if (u == -1) break;
        used[u] = true;
        if (trace[u] != -1) total += a[u][trace[u]];

        for (int v = 1; v <= n; v++) {
            if (a[u][v] && !used[v] && a[u][v] < d[v]) {
                d[v] = a[u][v];
                trace[v] = u;
            }
        }
    }

    cout << n - 1 << " " << total << endl;
    for (int v = 2; v <= n; v++)
        cout << trace[v] << " " << v << " " << a[trace[v]][v] << endl;

    return 0;
}
````

###  Test Case 1 - Do thi trong de bai
Input:
```
7 12
1 2 8
1 3 5
2 3 10
2 4 12
2 5 18
3 4 3
3 6 16
4 5 14
4 7 4
5 7 21
6 7 26
5 6 7
```
Output:
```
6 42
3 4 3
4 7 4
1 3 5
5 6 7
1 2 8
3 6 16
```
Giai thich:
Bat dau tu dinh 1, moi buoc chon canh nho nhat noi voi dinh moi chua chon. Tong trong so la 42.
