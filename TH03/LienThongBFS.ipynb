# Duyệt đồ thị bằng BFS từ đỉnh x

### Input
- Dòng 1: `n x`
  - `n`: số đỉnh (đánh số từ `1` đến `n`)
  - `x`: đỉnh bắt đầu duyệt BFS
- `n` dòng tiếp theo: dòng thứ `i` chứa **danh sách kề** của đỉnh `i` (các đỉnh cách nhau bởi dấu cách).  
  - Không có `-1` kết thúc, đọc hết dòng là hết danh sách kề của đỉnh đó.
  - Chương trình **không tự thêm cạnh ngược**, nên tùy đề bài mà đây có thể là đồ thị có hướng.

### Output
- Dòng 1: số lượng đỉnh đi được từ `x` (tức là kích thước thành phần liên thông/khả đạt từ x).
- Dòng 2: liệt kê các đỉnh theo **thứ tự duyệt BFS**.


````cpp
#include <iostream>
#include <vector>
#include <queue>
#include <sstream>
#include <string>
using namespace std;

int main() {
    int n, x;
    cin >> n >> x;      // số đỉnh và đỉnh bắt đầu 
    cin.ignore();       // bỏ ký tự xuống dòng còn sót

    vector<vector<int>> dsKe(n + 1);

    // đọc danh sách kề (đồ thị có hướng hoặc vô hướng tùy đề bài, không tự thêm cạnh ngược)
    for (int i = 1; i <= n; i++) {
        string line;
        getline(cin, line);
        stringstream ss(line);
        int v;
        while (ss >> v) {
            dsKe[i].push_back(v);
        }
    }

    // BFS
    vector<bool> daTham(n + 1, false);
    queue<int> hangdoi;
    vector<int> ketqua;

    daTham[x] = true;
    hangdoi.push(x);

    while (!hangdoi.empty()) {
        int u = hangdoi.front();
        hangdoi.pop();
        ketqua.push_back(u);

        for (int v : dsKe[u]) {
            if (!daTham[v]) {
                daTham[v] = true;
                hangdoi.push(v);
            }
        }
    }
    cout << ketqua.size() << "\n";
    for (int i = 0; i < (int)ketqua.size(); i++) {
        cout << ketqua[i];
        if (i < (int)ketqua.size() - 1) cout << " ";
    }
    cout << "\n";

    return 0;
}
````

### Test 1 — Đường thẳng đơn giản

**Input**
```text
4 1
2
3
4

```

Giải thích:

- Danh sách kề:
  - 1: 2
  - 2: 3
  - 3: 4
  - 4: (rỗng)
- Bắt đầu từ `1` → BFS lần lượt thăm: `1, 2, 3, 4`.

**Output**
```text
4
1 2 3 4
```

### Test 2 — Đồ thị có nhánh rẽ

**Input**
```text
5 1
2 3
4
5


```

Giải thích:

- Danh sách kề:
  - 1: 2, 3
  - 2: 4
  - 3: 5
  - 4: (rỗng)
  - 5: (rỗng)
- BFS từ `1`:
  - Hàng đợi ban đầu: [1]
  - Thăm 1 → đưa 2, 3 vào hàng đợi → [2, 3]
  - Thăm 2 → đưa 4 vào hàng đợi → [3, 4]
  - Thăm 3 → đưa 5 vào hàng đợi → [4, 5]
  - Thăm 4, rồi 5.
- Thứ tự duyệt: `1 2 3 4 5`.

**Output**
```text
5
1 2 3 4 5
```

### Test 3 — Có đỉnh không đạt được từ x

**Input**
```text
5 2
3
4


5

```

Giải thích:

- Danh sách kề:
  - 1: 3
  - 2: 4
  - 3: (rỗng)
  - 4: (rỗng)
  - 5: (rỗng)
- Bắt đầu từ `2` chỉ tới được `4`.  
  - `2 → 4`
- Các đỉnh 1, 3, 5 **không đạt được** từ 2.

**Output**
```text
2
2 4
```
