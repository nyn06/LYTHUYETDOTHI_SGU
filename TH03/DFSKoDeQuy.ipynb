# Duyệt đồ thị bằng DFS dùng ngăn xếp (phi đệ quy)

## Mô tả bài toán

- Input:
  - Dòng 1: `n x`
    - `n`: số đỉnh (đánh số từ `1..n`)
    - `x`: đỉnh bắt đầu duyệt DFS
  - `n` dòng tiếp theo: dòng thứ `i` chứa **danh sách kề** của đỉnh `i`, các đỉnh cách nhau bởi dấu cách.
    - Nếu có giá trị `-1` thì sẽ bị bỏ qua.
    - Đồ thị là **vô hướng**, nên khi đọc cạnh `i -> v` ta cũng thêm cạnh `v -> i`.

- Output:
  - Dòng 1: số lượng đỉnh đã duyệt được từ đỉnh bắt đầu `x`.
  - Dòng 2: các đỉnh theo thứ tự duyệt DFS (theo cách sử dụng ngăn xếp hiện tại).


````cpp
#include <iostream>
#include <vector>
#include <sstream>
#include <string>
#include <stack>
using namespace std;

int main() {
    int n, x;
    cin >> n >> x;
    cin.ignore();

    vector<vector<int>> dsKe(n + 1);

    // đọc danh sách kề
    for (int i = 1; i <= n; i++) {
        string line;
        getline(cin, line);
        stringstream ss(line);
        int v;
        while (ss >> v) {
            if (v != -1) {
                dsKe[i].push_back(v);
                dsKe[v].push_back(i); // vô hướng
            }
        }
    }

    vector<bool> daTham(n + 1, false);
    vector<int> ketqua;

    stack<int> st;
    st.push(x);
    daTham[x] = true;

    // DFS dùng ngăn xếp (phi đệ quy)
    while (!st.empty()) {
        int u = st.top();
        st.pop();
        ketqua.push_back(u);

        for (int v : dsKe[u]) {
            if (!daTham[v]) {
                daTham[v] = true;
                st.push(v);
            }
        }
    }

    cout << ketqua.size() << "\n";
    for (int i = 0; i < (int)ketqua.size(); i++) {
        cout << ketqua[i];
        if (i < (int)ketqua.size() - 1) cout << " ";
    }
    cout << "\n";

    return 0;
}
````


### Test 1 — Đồ thị là đường thẳng

**Input**
```text
4 1
2
1 3
2 4
3
```

Giải thích:

- Đồ thị: `1 - 2 - 3 - 4`
- Bắt đầu DFS từ đỉnh `1`.
- Thứ tự duyệt (theo ngăn xếp, với danh sách kề như trên): `1 2 3 4`.

**Output**
```text
4
1 2 3 4
```

### Test 2 — Đồ thị có 2 thành phần, bắt đầu từ 1 đỉnh trong thành phần thứ 2

**Input**
```text
5 4
2
1 3
2
5
4
```

Giải thích:

- Thành phần 1: `1 - 2 - 3`
- Thành phần 2: `4 - 5`
- Bắt đầu DFS từ `4`, chỉ duyệt được các đỉnh trong thành phần chứa `4`.
- Thứ tự duyệt có thể là `4 5` (tùy theo thứ tự danh sách kề).

**Output**
```text
2
4 5
```

### Test 3 — Đồ thị có nhánh rẽ

**Input**
```text
5 1
2 3
1 4
1 5
2
3
```

Giải thích:

- Đồ thị:
  - Từ `1` nối tới `2` và `3`
  - `2` nối tới `4`
  - `3` nối tới `5`
- Bắt đầu DFS từ `1`:
  - `1` → đẩy lần lượt `2`, `3` vào stack
  - Pop ra theo LIFO: sẽ duyệt `3` trước, rồi đến `5`, sau đó quay lại `2`, rồi `4`
- Một thứ tự duyệt hợp lệ: `1 3 5 2 4`

**Output**
```text
5
1 3 5 2 4
```
