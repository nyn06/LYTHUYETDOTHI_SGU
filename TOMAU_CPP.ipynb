# TOMAU_CPP - To mau do thi

````c++
#include <iostream>
#include <vector>
#include <fstream>
using namespace std;

const int MAX = 1001;
int n, m;
vector<int> adj[MAX]; // danh sach ke
int color[MAX];       // luu mau cua moi dinh

int main() {
    ifstream fin("TOMAU.INP");
    ofstream fout("TOMAU.OUT");

    fin >> n >> m;
    for (int i = 0; i < m; i++) {
        int u, v;
        fin >> u >> v;
        adj[u].push_back(v);
        adj[v].push_back(u);
    }

    for (int u = 1; u <= n; u++) {
        bool usedColor[MAX] = {false};
        for (int v : adj[u]) {
            if (color[v] != 0)
                usedColor[color[v]] = true;
        }
        int c = 1;
        while (usedColor[c]) c++;
        color[u] = c;
    }

    int maxColor = 0;
    for (int i = 1; i <= n; i++)
        if (color[i] > maxColor)
            maxColor = color[i];

    fout << maxColor << endl;
    for (int c = 1; c <= maxColor; c++) {
        for (int i = 1; i <= n; i++)
            if (color[i] == c)
                fout << i << " ";
        fout << endl;
    }

    fin.close();
    fout.close();
    return 0;
}
````

## Test Case 1 - Theo de bai
**Input:**
```
9 12
1 2
1 4
1 5
2 3
3 7
3 4
4 5
4 6
5 6
5 9
6 8
6 9
```
**Output:**
```
2
1 3 5 8
2 4 6 7 9
```
**Giai thich:** Can 2 mau de to, cac dinh ke nhau khong trung mau. Mau 1 cho 1 3 5 8, mau 2 cho 2 4 6 7 9.
