# Phân tích số mũ các thừa số nguyên tố trong N!
## Mô tả input / output

- **Input**: nhiều dòng, mỗi dòng là một số nguyên dương `N`.
- **Output**: với mỗi `N`, in một dòng gồm các số nguyên là số mũ của các số nguyên tố xuất hiện trong `N!`.

Ví dụ với `N = 5`:

- `5! = 120 = 2^3 * 3^1 * 5^1`
- Dãy số mũ là: `3 1 1`


````cpp
#include <iostream>
#include <vector>
using namespace std;

// Hàm kiểm tra số nguyên tố
bool snt(int x) {
    if (x < 2) return false;
    for (int i = 2; i * i <= x; i++) {
        if (x % i == 0) return false;
    }
    return true;
}

// Hàm tính số mũ của p trong N!
int demsomu(int N, int p) {
    int cnt = 0;
    while (N > 0) {
        N /= p;
        cnt += N;
    }
    return cnt;
}

int main() {
    int N;
    while (cin >> N) { // đọc nhiều test
        vector<int> result;
        for (int p = 2; p <= N; p++) {
            if (snt(p)) {
                int e = demsomu(N, p);
                result.push_back(e);
            }
        }
        // in kết quả
        for (int i = 0; i < (int)result.size(); i++) {
            cout << result[i];
            if (i + 1 < (int)result.size()) cout << " ";
        }
        cout << "\\n";
    }
    return 0;
}
````

## 3 test case mẫu

Dưới đây là 3 test case (input và output tương ứng):

### Test 1
**Input**
```
5
```

**Output**
```
3 1 1
```

---

### Test 2
**Input**
```
10
```

**Output**
Giải thích nhanh:

- Các số nguyên tố ≤ 10 là: 2, 3, 5, 7  
- Số mũ của 2 trong 10!:
  - ⌊10/2⌋ + ⌊10/4⌋ + ⌊10/8⌋ = 5 + 2 + 1 = 8  
- Số mũ của 3 trong 10!:
  - ⌊10/3⌋ + ⌊10/9⌋ = 3 + 1 = 4  
- Số mũ của 5 trong 10!:
  - ⌊10/5⌋ = 2  
- Số mũ của 7 trong 10!:
  - ⌊10/7⌋ = 1  

Vậy output:
```
8 4 2 1
```

---

### Test 3
**Input**
Nhiều test trên nhiều dòng:
```
3
6
10
```

**Output**
Giải thích:
- `3! = 6 = 2^1 * 3^1` → `1 1`
- `6! = 720 = 2^4 * 3^2 * 5^1` → `4 2 1`
- `10!` như trên → `8 4 2 1`

Vậy output:
```
1 1
4 2 1
8 4 2 1
```
