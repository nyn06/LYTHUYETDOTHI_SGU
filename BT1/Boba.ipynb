# Đếm tần suất lớn nhất của bộ ba ký tự liên tiếp trong xâu

- Đọc số lượng test `t`
- Với mỗi test, đọc một xâu ký tự `s` (không có khoảng trắng)
- Duyệt qua **tất cả các bộ ba ký tự liên tiếp** trong xâu `s`
- Đếm số lần xuất hiện của từng bộ ba (dùng `map<string, int>`)
- In ra **giá trị tần suất lớn nhất** trong các bộ ba của xâu đó

Nếu xâu có độ dài nhỏ hơn 3 thì vòng lặp không chạy và kết quả in ra sẽ là `0`.


````cpp
#include <iostream>
#include <string>
#include <map>
using namespace std;

int main()
{
    int t;
    cin >> t;
    while (t--) {
        string s;
        cin >> s;

        map<string, int> dem; // lưu số lần xuất hiện của từng bộ ba
        int maxx = 0;

        // duyệt qua tất cả bộ ba
        for (int i = 0; i + 2 < (int)s.size(); i++)
        {
            string boBa = s.substr(i, 3); // lấy ra 3 ký tự liên tiếp
            dem[boBa]++; // tăng tần suất
            if (dem[boBa] > maxx) {
                maxx = dem[boBa]; // cập nhật kết quả lớn nhất
            }
        }
        cout << maxx << "\n";
    }
    return 0;
}
````


### Test 1
**Input**
```text
1
ABCABC
```

Giải thích:

- Các bộ ba của `ABCABC`:
  - `ABC` (từ vị trí 0)
  - `BCA` (từ vị trí 1)
  - `CAB` (từ vị trí 2)
  - `ABC` (từ vị trí 3)
- Tần suất:
  - `ABC` xuất hiện 2 lần
  - `BCA` xuất hiện 1 lần
  - `CAB` xuất hiện 1 lần

**Output**
```text
2
```

  - Chỉ có một bộ ba: `ZZZ` xuất hiện 1 lần → in `1`

1
