# Kiểm tra dãy có phải hoán vị của 1..N hay không

Chương trình C++ dưới đây:

- Đọc số nguyên `N`.
- Đọc tiếp `N` số nguyên vào mảng `a`.
- Kiểm tra xem dãy `a` có phải là **hoán vị** của các số từ `1` đến `N` hay không.

Một dãy là hoán vị của `1..N` khi và chỉ khi:

1. Mỗi phần tử nằm trong đoạn từ `1` đến `N`.
2. Không có phần tử nào bị lặp lại (mỗi số từ `1` đến `N` xuất hiện đúng **1 lần**).

Nếu dãy thỏa điều kiện trên, in ra `YES`, ngược lại in `NO`.


````cpp
#include <iostream>
#include <vector>
using namespace std;

int main() {
    int N;
    cin >> N;
    vector<int> a(N);
    for (int i = 0; i < N; i++) {
        cin >> a[i];
    }

    vector<int> check(N + 1, 0); // check[i] = 1 nếu số i đã xuất hiện
    bool ok = true;

    for (int i = 0; i < N; i++) {
        if (a[i] < 1 || a[i] > N) {
            ok = false; // ngoài phạm vi
            break;
        }
        if (check[a[i]] == 1) {
            ok = false; // trùng số
            break;
        }
        check[a[i]] = 1; // đánh dấu đã gặp
    }

    if (ok) cout << "YES";
    else cout << "NO";

    return 0;
}
````


### Test 1
**Input**
```text
5
1 2 3 4 5
```

Giải thích:

- Các số đều trong đoạn `1..5`.
- Không có số nào trùng lặp.
- Dãy này chính là hoán vị của `1..5`.

**Output**
```text
YES
```
