# Đếm số lượng chữ số 0 tận cùng của N!

Đọc nhiều dòng (mỗi dòng một số `N`) và với mỗi `N` sẽ tính số lượng chữ số `0` ở cuối của `N!`.

Quy tắc:

- Số lượng số 0 tận cùng của `N!` chính là số mũ của 5 trong phân tích thừa số của `N!`.
- Công thức:
  - `ans = floor(N/5) + floor(N/25) + floor(N/125) + ...`

Chương trình chạy cho đến khi gặp **EOF**.


````cpp
#include <iostream>
using namespace std;

int main() {
    long long N;
    // đọc nhiều dòng cho tới khi hết file (EOF)
    while (cin >> N) {
        long long ans = 0;
        while (N > 0) {
            ans += N / 5;   // cộng số lượng bội của 5, 25, 125...
            N /= 5;
        }
        cout << ans << "\n";
    }
    return 0;
}
````


### Test 1
**Input**
```text
5
```

Giải thích:
- `5! = 120` → có **1 số 0** ở cuối.

**Output**
```text
1
```

---

### Test 2
**Input**
```text
10
```

Giải thích:
- `10! = 3 628 800` → có **2 số 0** ở cuối.
- Vì: `10/5 = 2` và `10/25 = 0`.

**Output**
```text
2
```

### Test 3
**Input**
```text
3
6
25
```

Giải thích nhanh:
- `3! = 6` → 0 số 0.
- `6! = 720` → 1 số 0.
- `25!` có `25/5 = 5` + `25/25 = 1` = **6** số 0.

**Output**
```text
0
1
6
```
